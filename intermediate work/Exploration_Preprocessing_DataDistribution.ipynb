```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read Train CSV
X_train_preprocessed = pd.read_csv("X_train_preprocessed.csv", index_col='row_index')
y_train_preprocessed = pd.read_csv("y_train_preprocessed.csv", index_col='row_index')
X_test_preprocessed = pd.read_csv("X_test_preprocessed.csv", index_col='row_index')
```

```python
# Box plots :

# Assuming dataset_X_train is your DataFrame
plt.figure(figsize=(15, 12))
sns.boxplot(data=X_train_preprocessed, orient="h", palette="Set2")
plt.title('Boxplot of Features')
plt.show()
```

```python
float_columns = X_train_preprocessed.select_dtypes(include=['float64']).columns.tolist()
```

```python


plt.figure(figsize=(20,20))

for c,var in enumerate(float_columns):
  plt.subplot(19,2,c+1)
  sns.kdeplot(data=X_train_preprocessed[float_columns], x=var, fill=var)
  plt.title(str(var))
  plt.xlabel("")
  plt.ylabel("")

plt.tight_layout()
```
